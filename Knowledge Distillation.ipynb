## Small Network Test

* Small model with distillation: 94.9% accuracy
* Small model without distillation: 91.4% accuracy

```python
tf.reset_default_graph()
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    teacher_model = model.TeacherModel(sess, dropout_prob=0.75, batch_size=128, learning_rate=0.001, temperature=5)
    teacher_model.load_model_from_file()
    
    stud_sess = tf.Session()
    student_model = model.StudentModel(stud_sess, batch_size=128, learning_rate=0.001, temperature=5)
    student_model.train(teacher_model)
```

```python
tf.reset_default_graph()
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    teacher_model = model.TeacherModel(sess, dropout_prob=0.75, batch_size=128, learning_rate=0.001, temperature=5)
    teacher_model.load_model_from_file()
    
    stud_sess = tf.Session()
    student_model = model.StudentModel(stud_sess, batch_size=128, learning_rate=0.001, 
                                       checkpoint_file='Student_raw', temperature=5, kd=False)
    student_model.train(teacher_model)
```
